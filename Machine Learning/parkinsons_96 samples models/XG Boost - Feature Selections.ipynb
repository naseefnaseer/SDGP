```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'parkinsons_train.csv', header=0)
```

```python
data.columns
```

```python
y = data.loc[:, 'status'].values
```

```python
cols_1 = ['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'Shimmer:APQ5',
       'MDVP:APQ', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']
```

```python
X = data[cols_1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 12)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
standard_X = StandardScaler()
standard_X.fit(X_train)
```

```python
X_train = standard_X.transform(X_train)
X_test = standard_X.transform(X_test)
X_train
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print(score)
```

```python
y_pred = model.predict(X_test)

print("Confusion Matrix")
print(confusion_matrix(y_test, y_pred))
print("\n")

print("Classfications Results")
print(classification_report(y_test, y_pred))

print("Accuracy Score")
print(accuracy_score(y_test, y_pred))
```
