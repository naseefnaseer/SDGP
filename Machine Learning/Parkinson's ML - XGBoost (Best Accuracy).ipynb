```python
X = data.loc[:, data.columns != 'Class information'].values[:, 1:]
y = data.loc[:, 'Class information'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size = 0.98, stratify = data['Class information'])
```

```python
from sklearn.preprocessing import StandardScaler
standard_X = StandardScaler()
```

```python
X_train = standard_X.fit_transform(X_train)
X_test = standard_X.fit_transform(X_test)
```

```python
X_train
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
print("Accuracy for training set -> ", end = "")
Y_hat = [round(yhat) for yhat in model.predict(X)]
print(accuracy_score(y, Y_hat)) # Full set accuracy
```

```python
Y_hat = [round(yhat) for yhat in model.predict(X_test)]
print("Accuracy for test set -> ", end = "")
print(accuracy_score(y_test, Y_hat)) # Test set accuracy
```
