```python
import joblib
decision_tree_model = joblib.load('model_decision_tree.pkl')
import sklearn
```

```python
print('The scikit-learn version is {}.'.format(sklearn.__version__))
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
pydot_data = StringIO()
export_graphviz(decision_tree_model, out_file=pydot_data,  
                filled=True, rounded=True,
                special_characters=True)
model_diagram = pydotplus.graph_from_dot_data(pydot_data.getvalue())  
Image(model_diagram.create_png())
```
